# Research Questions

In this step, we will address two specific research questions: 
1. Testing a claim from a Times of India article regarding the salary of fresh graduates in specific job roles.
2. Investigating whether gender influences specialization preferences using a chi-square test of independence.


```python
# Check the unique values in the Salary column
print("Unique values in Salary column:")
print(df['Salary'].unique())
```

```python
# Check unique values in the Designation column
print("Unique Designation values:")
print(df['Designation'].unique())
```

```python
import scipy.stats as stats

# Example DataFrame (df)
# df = pd.DataFrame({'Designation': [...], 'Salary': [...]})

# Specify the claimed salary range
lb = 250000  # Lower bound
ub = 300000  # Upper bound

# Filter data for specified job titles
job_titles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']

# Standardize job titles in the DataFrame
df['Designation'] = df['Designation'].str.strip().str.lower()

# Convert job_titles to lowercase for comparison
job_titles_lower = [title.lower() for title in job_titles]

# Filter the DataFrame
filtered_data = df[df['Designation'].isin(job_titles_lower)]

# Check if any data was found
if not filtered_data.empty:
    # Perform one-sample t-test on salary within the specified bounds
    t_statistic, p_value = stats.ttest_1samp(filtered_data['Salary'], lb)

    # Display the results
    print(f"T-statistic: {t_statistic}, P-value: {p_value}")

    # Interpret the p-value
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")

    # Check for average salary in the specified range
    average_salary = filtered_data['Salary'].mean()
    if lb <= average_salary <= ub:
        print(f"The average salary for specified designations is within the range: {average_salary}")
    else:
        print(f"The average salary for specified designations is not within the range: {average_salary}")
else:
    print("No data found for the specified job titles.")
```

```python
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

# Step 2: Perform the Chi-Square Test of Independence
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Step 3: Display the contingency table
print("Contingency Table:")
print(contingency_table)

# Step 4: Display the Chi-Square test results
print(f"\nChi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Step 5: Interpret the result
alpha = 0.05
if p < alpha:
    print("\nSince p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a significant relationship between Gender and Specialization.")
else:
    print("\nSince p-value is greater than 0.05, we fail to reject the null hypothesis.")
    print("There is no significant relationship between Gender and Specialization.")

# Step 6: Visualize the relationship using a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(contingency_table, annot=True, cmap='Blues')
plt.title('Heatmap of Gender vs Specialization')
plt.ylabel('Gender')
plt.xlabel('Specialization')
plt.show()
```

# Bonus Research Question:
## Is Job Location a Major Determinant of Salary?

```python
# Filter relevant columns
location_salary = df[['JobCity', 'Salary']].dropna()

# Check for unique job locations (you might want to limit the number of unique cities)
print("Unique job locations:", location_salary['JobCity'].nunique())

# Group data by city and filter cities with enough data points (at least 10 entries)
city_counts = location_salary['JobCity'].value_counts()
location_salary_filtered = location_salary[location_salary['JobCity'].isin(city_counts[city_counts >= 10].index)]

# Plotting the distribution of salaries by Job City
plt.figure(figsize=(15, 8))
sns.boxplot(x='JobCity', y='Salary', data=location_salary_filtered)
plt.title('Salary Distribution by Job Location')
plt.xlabel('Job Location')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()

# Performing One-Way ANOVA Test
city_groups = [location_salary_filtered[location_salary_filtered['JobCity'] == city]['Salary'] 
               for city in location_salary_filtered['JobCity'].unique()]

anova_city_result = stats.f_oneway(*city_groups)
print(f"One-Way ANOVA Test for Salary by Job Location: F-statistic = {anova_city_result.statistic}, p-value = {anova_city_result.pvalue}")

if anova_city_result.pvalue < 0.05:
    print("There is a significant difference in salaries across job locations.")
else:
    print("There is no significant difference in salaries across job locations.")
```
